#### Concept of argmax() and argmin()

```python
import numpy as np

a = np.array([[1,9,7],[2,5,0],[6,3,8]])
a
```

```python
a.argmin()
```

```python
a.argmax()
```

```python
a.flatten()
```

```python
a.argmin(axis=0)
```

```python
a.argmin(axis=1)
```

```python
a.argmax(axis=0)
```

```python
a.argmax(axis=1)
```
